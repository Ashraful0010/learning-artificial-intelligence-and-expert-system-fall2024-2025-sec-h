```python
def createAdjList(n):
    adj_list = [[] for _ in range(n)]
    return adj_list

def printAdjList(adj_list):
    for i in range(len(adj_list)):
        print(f"{i}: {adj_list[i]}")

f = open("graph.txt", "r")
v_e = f.readline().split()
n = int(v_e[0])
m = int(v_e[1])

adj_list = createAdjList(n)

print()

for i in range(m):
    l = f.readline().split()
    u = int(l[0])
    v = int(l[1])
    adj_list[u].append(v)
    adj_list[v].append(u)  # for undirected graph

printAdjList(adj_list)

print()

import queue

def dfs(adj_list, s, n):
    stack = queue.LifoQueue()
    visited = [0] * n
    distance=[0]*n
    visited[s] = 1
    distance[s] = 0
    stack.put(s)
    while not stack.empty():
        node = stack.get()
        print(node, end=" ")
        for i in adj_list[node]:
            if visited[i] == 0:
                visited[i] = 1
                distance[i]=distance[node]+1
                stack.put(i)

    print()
    print(distance)

dfs(adj_list, 1, n)
```
